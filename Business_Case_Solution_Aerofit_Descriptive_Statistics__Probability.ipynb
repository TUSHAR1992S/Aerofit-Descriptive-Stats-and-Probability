```python
#Age Column:
#1. Young: from 18(min) - 26
#2. Adults: from 27 - 35
#3. Middle_Aged: 36-45
#4. Senior :46 and above (max limit is 50 choosen value 51)
age_brange = [17,25,35,45,51]
age_blabel = ['Young', 'Adults', 'Middle_Aged', 'Seniors']
df_aero_tm['age_group'] = pd.cut(df_aero_tm['Age'],bins = age_brange,labels = age_blabel)
df_aero_tm.head()
```

#**Univariate Analysis**: Data Distribution and Outliners

Distribution of Products, Gender and Marital Status

* **Distribution of Products**

```python
prod_dist= df_aero_tm["Product"].value_counts().to_frame()
prod_dist = prod_dist.reset_index()
prod_dist.rename(columns = {'index':'Product_type','Product':'Product_Count'}, inplace=True)
prod_dist
```

```python
def tota_rev(df):
  if df['Product_type'] == 'KP281':
    return df['Product_Count'] * 1500
  elif df['Product_type'] == 'KP481':
    return df['Product_Count'] * 1750
  elif df['Product_type'] == 'KP781':
    return df['Product_Count'] * 2500
  else:
    return 0
```

```python
prod_dist['rev'] = prod_dist.apply(tota_rev, axis=1)
prod_dist
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('% Contribution of each Product')
plt.pie(x = prod_dist['Product_Count'], labels = prod_dist['Product_type'], autopct = '%.2f%%')

plt.subplot(1,2,2)
plt.title('% Contribution of each Product by Sale Revenue')
plt.pie(x = prod_dist['rev'], labels = prod_dist['Product_type'], autopct = '%.2f%%')
plt.show()
```

**Key Observation:**

* **Most Sales:**  highest number of units sold treadmill model is KP281 about 44% sale.
* **Sale By value:** All three models have approx equal contributions in terms of sales by revenue.

**7 (Q6.) With all the above steps you can answer questions like: What is the probability of a male customer buying a KP781 treadmill?**

```python
print('Conditional Probability')
print('-- '*10)
para_1 = df_aero_tm['Product'].unique()
para_2 = df_aero_tm['Gender'].unique()
for i in para_2:
  print()
  print(f'The conditional probability of a "{i}" purchasing :')
  for j in para_1:
    prob = len(df_aero_tm[(df_aero_tm['Gender']==i)&(df_aero_tm['Product']==j)])/len(df_aero_tm[df_aero_tm['Gender']==i])
    prob = np.round(prob * 100, 1)
    print(f"- '{j}' trademill is {prob} % ")
```
